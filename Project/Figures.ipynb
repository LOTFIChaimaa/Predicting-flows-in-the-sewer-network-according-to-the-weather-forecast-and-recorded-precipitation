### Analyse exploratoire de données

```python
#Les bibliothéques nécessaires
%matplotlib inline
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import xlrd
import seaborn as sns
```

#####  La figure suivant est un Histrogramme qui represente les précipitations en fonction de temps

```python
#J'ai utilisé la somme de précipitations pour toutes les stations

fig, axes = plt.subplots(figsize=(15,5))
precipitations = pd.read_excel("../Projet_v2/Excel_Sam.xlsx")

plt.bar(precipitations["Date"], precipitations["somme"], align='edge', width =0.03, linewidth=1)

plt.ylabel("Précipitations")
plt.xlabel("Date")
plt.title("Précipitations en fonction de temps")
plt.savefig('Précipitations_temps.png')
```

##### La figure suivant est un Histrogramme qui represente les deux premiers débits en fonction de temps

```python
fig, axes = plt.subplots(figsize=(15,5))
reg = pd.read_excel("../Projet_v2/debit_BCFI0010.xlsx")
debit1 = reg['BCFI0010']
debit2 = reg['BSFI0010']

plt.bar(reg['Date'], reg['BCFI0010'], align='edge', width =0.03, linewidth=1)
plt.bar(reg['Date'], reg['BSFI0010'], align='edge', width =0.03, linewidth=1)


plt.xlabel("Date")
plt.title("Débit BCFI0010 et BSFI0010 en fonction de temps")
plt.legend(['Débit BCFI0010', 'Débit BSFI0010'])
plt.savefig('Debit_1_2_temps.png')
```
